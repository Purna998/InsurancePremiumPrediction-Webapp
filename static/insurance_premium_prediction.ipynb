### Handle Categorical Columns

```python
df.head()
```

```python
df.sex.unique()
```

```python
df['sex']=df.sex.map({'female':0,'male':1})
```

```python
df.head()
```

```python
df.smoker.unique()
```

```python
df['smoker']=df.smoker.map({'no':0,'yes':1})
```

```python
df.head()
```

```python
df.region.unique()
```

```python
df['region']=df.region.map({'southwest':1,'southeast':2,'northwest':3,'northeast':4})
```

```python
df.head()
```

```python
df.info()
```

### Split the dataset(x-independent and y-dependent on x)


```python
df.columns
```

```python
x=df.drop('expenses',axis=1)
```

```python
x.head()
```

```python
y=df.expenses
```

```python
y.head()
```

```python
y.shape
```

### Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
df.shape
```

```python
x_test.shape
```

### Model Training

### Linear Regression

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_predict=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
score1=r2_score(y_test,y_predict)
accuracy1=score1*100
```

```python
print(f"{accuracy1:.2f}%")
```

### Support vector Regressor

```python
#support vector regressor
from sklearn.svm import SVR
```

```python
svm=SVR()
```

```python
svm.fit(x_train,y_train)
```

```python
y_predict=svm.predict(x_test)
from sklearn.metrics import r2_score
score2=r2_score(y_test,y_predict)
accuracy2=score2*100
print(f"{accuracy2:.2f}%")
```

### RandomForest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor() 
```

```python
rf.fit(x_train,y_train)
```

```python
y_predict=rf.predict(x_test)
from sklearn.metrics import r2_score
score3=r2_score(y_test,y_predict)
accuracy3=score3*100
print(f"{accuracy3:.2f}%")
```

### Final Model is RandomForest Regressor

```python
rf.fit(x_train,y_train)
```
